# Lesson 1

## What is Data Engineering?

**Moving data from one place to another to make data available for the end-user, while doing some transformation and aggregation along the way (e.g. from a huge set of files into a database of some sort)

* It plays a vital role in storing (design efficient ways to store data), organizing (running pipelines to transform and move data from one place to the next) and preparing data (cleaning and summarizing it in a way that helps the end-user).
* The pipelines we build and databases we design are the foundation of all other analytics, data science, and AI.
* First and most important step in the digital transformation process.

* Move/Store: Storing data in databases and warehouses and using pipelines to move it from one place to the next 
* Explore/Transform: Checking for data accuracy, cleaning and preparing data for use
* Aggregate/Label: Summarizing and measuring what's already happened. Usually this is as far as companies get

## Learning Best Practices

* Take notes, such as screenshots and then notes beneath
* Write out code. It takes repetition to learn it
* Internalize content by 1) utilizing the community and 2) googling errors and reading documentation when stuck
* Help others; it's the best way to learn.
* Learn a few selected skills first (SQL, Python, Airflow, Scala, Redshift). Better to have a lot of expertise in a few big, important tools than it is to learn a bit of a lot of tools.
* When startng out, focus on these key questions: 
    * How am I going to organize my data pipelines?
    * How am I going to run big analyses that work on a lot of data?
